```python
import pandas as pd
import pickle
import numpy as np
```

```python
data = pd.read_csv("NegociationsTraffic.csv")
columnsdrop=['Bytes','Packets']
data
```

```python
if 'Unnamed: 0' in data.columns:
  data.drop(['Unnamed: 0'],axis=1,inplace=True)
data=data.drop(columns=columnsdrop,axis=1)
data=data.drop("Action",axis=1)
```

```python
df=data.drop('"Source Port"',axis=1)
```

```python
df
```

```python
with open('dstportdict.pkl', 'rb') as f:
    dstportdict = pickle.load(f)
with open('natsrcportdict.pkl', 'rb') as f:
    natsrcportdict = pickle.load(f)
with open('natdstportdict.pkl', 'rb') as f:
    natdstportdict = pickle.load(f)

# Appliquer la transformation en utilisant le dictionnaire chargé
df['"Destination Port"']=df['"Destination Port"'].map(dstportdict)
df['"NAT Source Port"']=df['"NAT Source Port"'].map(natsrcportdict)
df['"NAT Destination Port"']=df['"NAT Destination Port"'].map(natdstportdict)
```

```python
with open('countfrequencyencoderGAN.pkl', 'rb') as f:
    countfreqGAN = pickle.load(f)
```

```python
df = countfreqGAN.transform(df)
```

```python
df
```

```python
with open('standardscalerGAN.pkl','rb') as f:
    scaler=pickle.load(f)
```

```python
num=['"Bytes Sent"','"Bytes Received"','"Elapsed Time (sec)"',
                   "pkts_sent","pkts_received"]
df[num]=scaler.transform(df[num])
```

```python
dfsansNan = df[~df.isin([np.nan, np.inf, -np.inf]).any(axis=1)]
```

```python
data[df.isin([np.nan, np.inf, -np.inf]).any(axis=1)]
```

```python
dfsansNan
```

```python
import tensorflow as tf
allow_model = tf.keras.models.load_model("allow_detection.h5")
# Première étape: identification des données "allow"
allow_probs = allow_model.predict(dfsansNan)
rmse = np.sqrt(np.sum((dfsansNan - allow_probs)**2, axis=1) / dfsansNan.shape[1])

# Créer un masque pour les observations dont la "RMSE" est inférieure au seuil
allow_mask = rmse > 0.08502309105681057
```

```python
compteur_false = allow_mask.value_counts()[False]
compteur_false
```

```python
with open('DTC.pkl', 'rb') as f:
    Decision_tree_classifier = pickle.load(f)
```
